```python
base_url = "https://www.vezeeta.com/"  # Base URL for appending doctor profiles

for url in urls:
    try:
        # Introduce a delay to avoid detection
        time.sleep(random.uniform(1, 3))

        # Send request with headers
        response = requests.get(url, headers=headers)
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find all spans with the specific itemtype
            spans = soup.find_all("span", itemtype="schema.org/Physician")
            
            # If no spans are found, continue to the next URL
            if not spans:
                print(f"No doctors found on {url}")
                continue

            # Extract and print information from each span
            for span in spans:
                # Example: Extracting text or nested elements
                doctor_name = span.text.strip() if span.text else "No Name Found"
                print(f"Doctor Name: {doctor_name}")
                
                # Extract the relative doctor URL
                doctor_link = span.find("a")['href'] if span.find("a") else "No Link Found"
                
                # Create the full URL
                full_doctor_url = base_url + doctor_link.lstrip('/') if doctor_link != "No Link Found" else "No URL Found"
                print(f"Doctor Full URL: {full_doctor_url}")
                
                # Append the full URL to the list
                url_list.append(full_doctor_url)
        else:
            print(f"Failed to retrieve content from {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred while processing {url}: {str(e)}")
```
